```python
df['Time_Orderd_original'] = df['Time_Orderd']
df['Time_Order_picked_original'] = df['Time_Order_picked']
```

```python
def convert_excel_time(value):
    try:
        value = str(value)

        if ':' in value:
            return value

        decimal_time = float(value)
        total_minutes = round(decimal_time * 24 * 60)

        hours = total_minutes // 60
        minutes = total_minutes % 60

        if hours == 24:
            hours = 0

        return f"{hours:02d}:{minutes:02d}"

    except:
        return value
```

```python
df['Time_Orderd'] = df['Time_Orderd'].apply(convert_excel_time)
df['Time_Order_picked'] = df['Time_Order_picked'].apply(convert_excel_time)
```
